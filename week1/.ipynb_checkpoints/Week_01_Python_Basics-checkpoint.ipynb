```python
import pandas as pd
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("house_prices.csv")

# QUESTION 1: Load a CSV using pandas and print first 10 rows

print("--- QUESTION 1: First 10 Rows of Dataset ---")
print(df.head(10))

# QUESTION 2: Split dataset into train/test using sklearn
X = df[["Area_SqFt"]]
Y= df["Price_INR"]

X_train , X_test, Y_train, Y_test =train_test_split( X, Y , test_size=0.2 )

print("\n--- QUESTION 2: Data Split Check ---")
print(f"Training set size: {X_train.shape[0]} rows")
print(f"Testing set size: {X_test.shape[0]} rows")

# QUESTION 3: Train a Linear Regression model and check accuracy

model=LinearRegression()
model.fit(X_train, Y_train)

print("\n--- QUESTION 3: Model Evaluation ---")

score= model.score( X_test, Y_test )
print(f"\n Model R-squared Accuracy Score:{score *100}")

print("\n--- QUESTION 4: House Price Predictor Tool ---")

user_area = float(
    input("Enter house area in sq ft (e.g., 1500): ")
)  # Prompts you to type inside browser cell
user_input_df = pd.DataFrame([[user_area]], columns=['Area_SqFt'])

# Reshape input to 2D array as required by sklearn

predicted_price = model.predict(user_input_df)
print(f"➡️ Predicted House Price: INR {predicted_price[0]:,.2f}")


```
